## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.metrics import accuracy_score
```

### Importing the dataset

```python
df = pd.read_csv("creditcard.csv")
display(df.shape)
df.head()

#legit transaction -> class=0
#fraudulent transaction -> class=1
```

```python
df.info()  # to check the presence of null cells and datatype of various columns
```

```python
df.describe()  # to get an idea about the columns in general which will be useful later
```

```python
df["Class"].value_counts()
```

#### as you can see above the data is highly unbalanced, which is expected as most transactions would be legit (Class-->0)

### Creating separate dataframes for analysis

```python
legit_transactions = df[df["Class"] == 0]
fraudulent_transactions = df[df["Class"] == 1]

display(legit_transactions.head())
display(legit_transactions.shape)
display(fraudulent_transactions.head())
fraudulent_transactions.shape
```

```python
# Comparing legit and fraudulent transaction amounts

display(legit_transactions["Amount"].describe())
fraudulent_transactions["Amount"].describe()
```

### **Applying Under Sampling method to create a subset of legit_transactions so that we could build a balanced dataframe for machine learning

```python
# taking a random sample from legit_transactions

legit_sample = legit_transactions.sample(n=492)
```

```python
# concatinating the fraudulent_transactions and legit_sample to build final dataframe for machine learning

balanced_credit_card_data = pd.concat([legit_sample, fraudulent_transactions], axis=0)  # axis=0 means add row-wise
display(balanced_credit_card_data.head())
balanced_credit_card_data.tail()
```

```python
balanced_credit_card_data["Class"].value_counts()
```

```python
# Comparing the mean values of balanced_credit_card_data and original credit card data

display(balanced_credit_card_data.groupby("Class").mean())
df.groupby("Class").mean()
```

#### as you can see above the balanced_credit_card_data is quite similar to the original dataset, therefore, it could be used for training our machine learning model

## Building machine learning model

```python
#splitting the data into features and target

X = balanced_credit_card_data.drop(columns="Class")
y= balanced_credit_card_data["Class"]

display(X.head())
y
```

#### splitting the data into testing and training datasets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
# stratify=True in order to split the data evenly b/w the two classes, random_state=2 to reproduce the code
```

```python
print(X_train.shape, X_test.shape)
```

#### importing Logistic regression model for training our data

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train,y_train)
```

#### checking the accuracy of our model

```python
from sklearn.metrics import accuracy_score
```

```python
# accuracy score of training data

X_train_predictions = model.predict(X_train)
train_accuracy = accuracy_score(X_train_predictions, y_train)
train_accuracy
```

```python
# accuracy score of testing data

X_test_predictions = model.predict(X_test)
test_accuracy = accuracy_score(X_test_predictions, y_test)
test_accuracy
```
